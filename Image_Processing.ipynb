# Image Processing using Python

```python
import cv2
import numpy as np
#import pandas as pd

image = cv2.imread(r'C:/Users/Arvind/Desktop/Images/heic2007a.jpg')
row,col,plane = image.shape
print('original image dimensions:')
print(row,col,plane)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
row1,col1 = gray.shape
print('gray image dimensions:')
print(row1,col1)

image2 = cv2.imread(r'C:/Users/Arvind/Desktop/Images/img1.png')
row2,col2,plane2 = image2.shape
print('To merge image dimensions:')
print(row,col,plane)

dsize = (256,256)

# resize image
output = cv2.resize(image, dsize)
output1 = cv2.resize(image2, dsize)
      
# add or blend the images
output_b = cv2.addWeighted(output, 1, output1, 1, 0.0)

#gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
row1,col1,plane= output.shape
print('Resized image dimensions:')
print(row1,col1)
  
cv2.imshow('Resized image', output)
cv2.imshow('Original image',image)
cv2.imshow('Gray image', gray)

red_image = np.zeros((row,col,plane),np.uint8)

red_image[:,:,2]=image[:,:,2]
cv2.imshow('Red image', red_image)

green_image = np.zeros((row,col,plane),np.uint8)

green_image[:,:,1]=image[:,:,1]
cv2.imshow('Green image', green_image)

blue_image = np.zeros((row,col,plane),np.uint8)

blue_image[:,:,0]=image[:,:,0]
cv2.imshow('Blue image', blue_image)

cv2.imshow('Title image', image2)

cv2.imshow('Blended image', output_b)

cv2.waitKey(0)

print('')

isWritten1 = cv2.imwrite(r'C:/Users/Arvind/Desktop/Images/heic2007a_gray.jpg', gray)
if isWritten1:
    print('The gray image is successfully saved.')
    
isWritten2 = cv2.imwrite(r'C:/Users/Arvind/Desktop/Images/heic2007a_resized.jpg', output)
if isWritten2:
    print('The resized image is successfully saved.')
    
isWritten3 = cv2.imwrite(r'C:/Users/Arvind/Desktop/Images/heic2007a_red.jpg', red_image)
if isWritten3:
    print('The red image is successfully saved.')
    
isWritten4 = cv2.imwrite(r'C:/Users/Arvind/Desktop/Images/heic2007a_blue.jpg', blue_image)
if isWritten4:
    print('The blue image is successfully saved.')
    
isWritten5 = cv2.imwrite(r'C:/Users/Arvind/Desktop/Images/heic2007a_green.jpg', green_image)
if isWritten5:
    print('The green image is successfully saved.')
    
isWritten6 = cv2.imwrite(r'C:/Users/Arvind/Desktop/Images/heic2007a_blended.jpg', output_b)
if isWritten6:
    print('The blended image is successfully saved.')
    
cv2.destroyAllWindows()
```
